```
from keras.preprocessing.image import ImageDataGenerator,array_to_img,img_to_array,load_img
```

```
datagen=ImageDataGenerator(rotation_range=40,width_shift_range=0.2,height_shift_range=0.2,shear_range=0.2,zoom_range=0.4,horizontal_flip=True,fill_mode='nearest')
```

```
img=load_img('/content/drive/My Drive/Deep Learning/single_pred/dog.4605.jpg')
```

```
img
```

```
X=img_to_array(img)
```

```
X
```

```
X=X.reshape((1,)+X.shape)
```

```
X.shape
```

```
X
```

```
i=0
for batch in datagen.flow(X,batch_size=1,save_to_dir='/content/drive/My Drive/Deep Learning/single_pred/sample', save_prefix='dog',save_format="jpeg"):
  i+=1
  if i>20:
    break
```
